```python
# Import Dependecies

import pandas as pd
from sqlalchemy import create_engine
import requests
from pprint import pprint
import json

```

```python
# Read in 2 dataset urls for 2018-2019 years from github as json files
club_url='https://raw.githubusercontent.com/openfootball/football.json/master/2018-19/en.1.clubs.json'
match_url = 'https://raw.githubusercontent.com/openfootball/football.json/master/2018-19/en.1.json'

# Read them in
club_response=requests.get(club_url)
match_response=requests.get(match_url)

# Turn them to json files

club_json=club_response.json()
match_json=match_response.json()


```

```python
# Check how many 'Matchday' in 'rounds' key, and how many matches in that day
rounds=match_json['rounds']
matches=match_json['rounds'][0]['matches']

print(len(rounds))
print(len(matches))
# So we have 38 rounds and 10 matches in each round, so total we have 380 entries for the DataFrame
ft=(matches[0]['score']['ft'])
```

```python
# Create DataFrame for club_json

club_df=pd.DataFrame(club_json['clubs'])


# Create lists for match_json
match_list=[]
date_list=[]
team1_list=[]
team2_list=[]
score_list=[]
score1=[]
score2=[]

# Create nested for loop to take out the desired data
for x in range(len(rounds)):
    match_list.append(match_json['rounds'][x]['matches'])
    for y in range(len(matches)):
        date_list.append(match_json['rounds'][x]['matches'][y]['date'])
        team1_list.append(match_json['rounds'][x]['matches'][y]['team1'])
        team2_list.append(match_json['rounds'][x]['matches'][y]['team2'])
        score_list.append(match_json['rounds'][x]['matches'][y]['score'])
        score1.append(match_json['rounds'][x]['matches'][y]['score']['ft'][0])
        score2.append(match_json['rounds'][x]['matches'][y]['score']['ft'][1])

    
# Double check the len of each list to confirm the correct total entries
print(len(match_list))
print(len(score_list))
print(len(team1_list))
```

```python
# Create the DataFrame for the match_json
match_df=pd.DataFrame()
match_df['date']=date_list
match_df['team1']=team1_list
match_df['team2']=team2_list
match_df['team_1_score']=score1
match_df['team_2_score']=score2


```
